```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
import random
import numpy as np
```

```python
def entropy(y):
    e = None
    value_counts = y.value_counts()
    proportions = value_counts / len(y)
    e = np.sum((-1*proportions * np.log2(proportions)))
    return e

def gain(y,x):
    weighted_entropies = x.groupby(x).apply(lambda x_group: entropy(y[x_group.index]) * len(x_group) / len(x))
    weighted_average_entropy_x = weighted_entropies.sum()

    return entropy(y) - weighted_average_entropy_x


def gain_ratio(y,x):
    g = gain(y,x)
    if entropy(y) == 0:
        return 0
    return g/entropy(y)

def select_split(X,y):
    col = None
    gr = 0.0
    maxg = 0.0

    for c in X.columns:
        if y.name == c:
            continue
        en = entropy(X[c])
        tg = gain(y, X[c])
        tgr = gain_ratio(y, X[c])
        if maxg < tg:
            maxg = tg
            gr = tgr
            col = c

    if maxg == 0.0:
        return None, y.mode().iloc[0]

    return col, gr

def make_tree(X, y):
    tree = {}
    col, gr = select_split(X, y)

    if col is None:
        if(len(y.mode()) > 1):
            return y.iloc[0]
        return y.mode().iloc[0]

    tree[col] = {}

    for value in X[col].unique():

        X_subset = X[X[col] == value].drop(columns=col)
        y_subset = y[X[col] == value]
        tree[col][value] = make_tree(X_subset, y_subset)

    return tree

def make_prediction(tree, X_test):
    predictions = []

    for _, row in X_test.iterrows():
        node = tree

        # Traverse the decision tree until we reach a leaf node
        while isinstance(node, dict):
            feature = next(iter(node))
            value = row[feature]

            if value not in node[feature]:
                # If the feature value is not present in the decision tree, return a default prediction (e.g., 0.0)
                predictions.append(0.0)
                break

            node = node[feature][value]

        # If the current node is a leaf node, add its prediction to the list
        if not isinstance(node, dict):
            predictions.append(node)

    return np.array(predictions)
```

```python
def get_learner(X,y,max_depth=10):
    return DecisionTreeClassifier(max_depth=max_depth).fit(X,y)
```

```python
def resample(X,t):
    assert len(X) == len(t)
    indexes = []
    for i in range(len(X)):
        ix_choice = random.choice(X.index)
        indexes.append(ix_choice)
    return X.copy().loc[indexes],t.copy().loc[indexes]
```

```python
from sklearn.model_selection import train_test_split
```

```python
def make_trees(X,y,ntrees=2,max_depth=10):
    trees = []
    for i in range(ntrees):
        X_train_sample, y_train_sample = resample(X, y)
        t = get_learner(X_train_sample,y_train_sample)
        trees.append(t)
    return trees
```

```python
def get_learner_regression(X,y,max_depth=10):
    return DecisionTreeRegressor(max_depth=max_depth).fit(X,y)
```

```python
team_stats= {}
for idx, row in df_gdp.iterrows():
    home_team = row['home_team']
    away_team = row['away_team']
    winner = row['winner']
    home_score = row['home_score']
    away_score = row['away_score']

    # Update home team stats
    if home_team not in team_stats:
        team_stats[home_team] = {'wins': 0, 'total_games': 0,'point_diff': 0}

    team_stats[home_team]['total_games'] += 1
    team_stats[home_team]['wins'] += winner
    team_stats[home_team]['point_diff'] += home_score - away_score

    # Update away team stats
    if away_team not in team_stats:
        team_stats[away_team] = {'wins': 0, 'total_games': 0,'point_diff': 0}

    team_stats[away_team]['total_games'] += 1
    team_stats[away_team]['wins'] += (1 - winner)
    team_stats[away_team]['point_diff'] += away_score - home_score
#team_stats

df_gdpct = pd.DataFrame.from_dict(team_stats, orient='index')
df_gdpct.reset_index(inplace=True)
df_gdpct.columns = ['country', 'wins', 'total_games', 'point_diff']
df_gdpct['win_loss_ratio'] = df_gdpct['wins'] / df_gdpct['total_games']
```
